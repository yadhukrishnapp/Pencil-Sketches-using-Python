# Importing the Libraries

```python
import cv2 # pip install opencv-python
import matplotlib.pyplot as plt # pip install matplotlib
%matplotlib inline
```

```python
def show(title, img, cmap = 'RdBu'):
    plt.imshow(img, cmap = cmap)
    plt.title(title)
```

# Image Loading

```python
image = plt.imread("C:/Project/Image to Pencil Sketch/tiger.jpg")
show("Tiger", image, 'RdBu')
```

# Converting the Input Image into Grayscale Image

```python
grayscale_image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
show("Gray Image", grayscale_image, 'gray')
```

# Inverting the Grayscale Image

```python
inverted_image = 255 - (grayscale_image)
show("Inverted Image", inverted_image, 'gray')
```

# Blurring the Inverted Image

```python
blurred_image = cv2.GaussianBlur(inverted_image, (21, 21), 0)
show("Blurred Image", blurred_image, 'gray')
```

# Converting the Blurred Image into Pencil Sketch

```python
inverted_blurred = 255 - (blurred_image)
pencil_sketch = cv2.divide(grayscale_image, inverted_blurred, scale = 256.0)
show("Pencil Sketch", pencil_sketch, 'gray')
```
